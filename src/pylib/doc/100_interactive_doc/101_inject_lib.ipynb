This file provides examples on how to use the inject_lib function.


```python
# imports and preparations
from pylib import cli
import tempfile
from pathlib import Path

# lets enable autoreload so we can use these scrips for interactive src-debugging.
%load_ext autoreload
%autoreload 2
```

```python
# print the help of the function
# While this file calls the cli via the cli.run function, all command are also available from
# outside python via the shell. Use help(cli.run) for details.

cli.run(["inject-lib", "-h"])
```

```python
# lets create a tempory folder and inject into it.
td = tempfile.TemporaryDirectory(ignore_cleanup_errors=True)

root = Path(td.name)
cli.run(["inject-lib", str(root)])
```

```python
# remove the temp dir from above
td.cleanup()
```
